## Space Reduction

This notebook frees up spaces from in all specified folders (config.yaml).

```python
import os
import time
import yaml
import datetime
```

```python
def remove_flat_files(config_file):
    with open(config_file, 'r') as f:
        config_data = yaml.safe_load(f)

    current_time = time.time()

    for config in config_data:
        folder_path = config.get('path')
        age_threshold_days = config.get('age_threshold_days')
        if not folder_path or not age_threshold_days:
            print("Error: Invalid configuration data. Skipping this entry.")
            continue

        # Convert days to seconds using timedelta
        age_threshold_seconds = datetime.timedelta(days=age_threshold_days).total_seconds()

        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                modified_time = os.path.getmtime(file_path)
                age = current_time - modified_time

                if age > age_threshold_seconds:
                    os.remove(file_path)
                    modified_date = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(modified_time))
                    print(f"Deleted file: {file_path} (Modified Date: {modified_date})")

    print("Done processing all folders.")

remove_flat_files('config.yaml')
```
