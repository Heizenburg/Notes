```python
import pandas as pd
```

```python
def read_large_file(file_path):
    chunk_size = 1_000_000 
    df = pd.concat(pd.read_csv(file_path, chunksize=chunk_size))
    return df
```

```python
file_path = r'R:\RawData\PNP SAP\Tiger Brands_Tiger_Consumer_Brands_WeeklyExtract_20230717_20230723.csv'
df = read_large_file(file_path)
```

```python
file_path = r'R:\RawData\PNP SAP\Tiger_Consumer_Brands_WeeklyExtract_20230710_20230716.csv'
df_compare = read_large_file(file_path)
```

```python
columns_df1 = df.columns.tolist()
columns_df2 = df_compare.columns.tolist()
```

```python
max_length = max(len(columns_df1), len(columns_df2))
columns_df1 += [""] * (max_length - len(columns_df1))
columns_df2 += [""] * (max_length - len(columns_df2))

compared_df = pd.DataFrame({"Last Week": columns_df2, "This Week": columns_df1})
```
